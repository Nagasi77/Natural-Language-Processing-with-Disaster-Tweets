```python
import pandas as pd

jalur = '/kaggle/input/nlp-getting-started/'
df_latih = pd.read_csv(jalur + 'train.csv')
df_uji = pd.read_csv(jalur + 'test.csv')

print(df_latih.shape)
print(df_latih.head())
```

```python
print(df_latih['target'].value_counts())
print(df_latih.isnull().sum())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
mesin_kata = CountVectorizer()
X_latih = mesin_kata.fit_transform(df_latih['text'])
X_uji = mesin_kata.transform(df_uji['text'])
y_latih = df_latih['target']
print(X_latih.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_latih, y_latih)
print("Pelatihan selesai")
```

```python
tebakan = model.predict(X_uji)
df_kumpul = pd.DataFrame()
df_kumpul['id'] = df_uji['id']
df_kumpul['target'] = tebakan
df_kumpul.to_csv('submission.csv', index=False)
print('selesai')
```

```python
import re 

def bersihkan(teks): 
    teks = teks.lower() 
    teks = re.sub(r'http\S+', '', teks) 
    teks = re.sub(r'[^a-z]', ' ', teks) 
    return teks 

df_latih['text_bersih'] = df_latih['text'].apply(bersihkan) 
df_uji['text_bersih'] = df_uji['text'].apply(bersihkan) 
print(df_latih[['text', 'text_bersih']].head()) 
```

```python
X_latih_bersih = mesin_kata.fit_transform(df_latih['text_bersih'])
X_uji_bersih = mesin_kata.transform(df_uji['text_bersih'])
model.fit(X_latih_bersih, y_latih)
tebakan_baru = model.predict(X_uji_bersih)

df_kumpul_baru = pd.DataFrame()
df_kumpul_baru['id'] = df_uji['id']
df_kumpul_baru['target'] = tebakan_baru
df_kumpul_baru.to_csv('submission_bersih.csv', index=False)
print("Selesai")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd

mesin_tfidf = TfidfVectorizer()
X_latih_tfidf = mesin_tfidf.fit_transform(df_latih['text_bersih'])
X_uji_tfidf = mesin_tfidf.transform(df_uji['text_bersih'])

model_nb = MultinomialNB()
model_nb.fit(X_latih_tfidf, y_latih)
tebakan_nb = model_nb.predict(X_uji_tfidf)

df_kumpul_nb = pd.DataFrame()
df_kumpul_nb['id'] = df_uji['id']
df_kumpul_nb['target'] = tebakan_nb
df_kumpul_nb.to_csv('submission_nb.csv', index=False)
print("Selesai Naive Bayes Lengkap")
```
